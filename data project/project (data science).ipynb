```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression



# Extract the independent variables X and the dependent variable y
X = dataset.iloc[:, 1:25]
y = dataset.iloc[:, -1]
# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=44, shuffle=True)

# Create a logistic regression object and fit the training data
logreg = LogisticRegression(penalty='l1', solver='liblinear')
logreg.fit(X_train, y_train)

# Print the accuracyscore for the training and test sets
print('Logistic Regression Train Score is : ', logreg.score(X_train, y_train))
print('Logistic Regression Test Score is : ', logreg.score(X_test, y_test))

# Predict the target variable for the test data
y_pred = logreg.predict(X_test)

# Print the predicted and actual values for the first 10 samples in the test set
print('Predicted Value for Logistic Regression is : ', y_pred[:10])
print('Actual Value for Logistic Regression is : ', y_test[:10])
# Calculate the accuracy of the predictions
accuracy = logreg.score(X_test, y_test)

# Print the accuracy
print('Accuracy of the Logistic Regression model is : ', accuracy)
```
